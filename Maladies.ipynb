```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "http://www.allmusic.com/blog/post/classical-composers-and-their-maladies/"
headers = {'User-agent': 'Mozilla/5.0'}
```

```python
page = requests.get(url, headers=headers)
```

```python
soup = BeautifulSoup(page.content)
```

```python
print(soup.prettify())
```

```python
maladies = [maladies.string for maladies in soup.find_all('strong')]
```

```python
print maladies
```

```python
composers = [composers.string for composers in soup.find_all('a')]
```

```python
print(composers)
```

```python
composers = composers[79:145]
```

```python
print composers
```

```python
composers_clean = filter(None, composers)
composers_clean.remove('Eighth Symphony')
composers_clean.remove('Das Lied von der Erde,')
composers_clean.remove('Ninth')
composers_clean.remove('Ninth Symphony.')
composers_clean.remove('Tenth,')
composers_clean.remove('Carmen,')
composers_clean.remove('Moses und Aron,')
composers_clean.remove('Fanny Mendelssohn,')
composers_clean.remove('Antonio Salieri.')
```

```python
df = pd.DataFrame({'composers':composers_clean})
```

```python
df['composers'] = df['composers'].str.replace("\\.|,","")
df['composers'] = df['composers'].str.encode('utf-8')
```

```python
df
```

```python
maladies_list = [[maladies[0]],[maladies[1]],[maladies[2]],[maladies[3]]*5,[maladies[4]],[maladies[5]],[maladies[7]]*13, [maladies[8]],[maladies[9]],[maladies[11]], [maladies[13]]*12,[maladies[14]],[maladies[15]]*10]
```

```python
print(maladies_list)
```

```python
maladies_list = [item for sublist in maladies_list for item in sublist]
```

```python
len(maladies_list)
```

```python
df['maladies'] = maladies_list
```

```python
df
```

```python
df.loc[len(df)]=['Ludwig van Beethoven','Deafness']
df.loc[len(df)]=['Anton Bruckner','Obsessive-Compulsive Disorder']
df.loc[len(df)]=['Wolfgang Amadeus Mozart','Rheumatic Fever']
df.loc[len(df)]=['Ludwig van Beethoven','Syphilis']
df.loc[len(df)]=['Wolfgang Amadeus Mozart','Syphilis']
df.loc[len(df)]=['Robert Schumann','Syphilis']
```

```python
df
```

```python
df['maladies'] = df['maladies'].astype('category')
df['composers'] = df['composers'].astype('category')
```

```python
df.to_csv("data/composer-maladies.csv")
```
